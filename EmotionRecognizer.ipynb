```python
import cv2
from deepface import DeepFace
```

```python
img = cv2.imread("Resources/happy_girl.jpg")   
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img) ##by default opencv is based on BGR
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
predictions = DeepFace.analyze(img)
```

```python
predictions['emotion']
```

```python
predictions['dominant_emotion']
```

```python
faceCascade = cv2.CascadeClassifier('haarcascades/haarcascades/haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray, 1.1, 4)

for (x, y, w, h) in faces: 
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
cv2.putText(img, predictions['dominant_emotion'], (5,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 1, cv2.LINE_4) ;
```
